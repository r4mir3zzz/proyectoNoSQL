```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("/content/Label.DatosCorregidos.csv")
```

```python
df.head()
```

```python
specific_columns = ["Delito", "Provincia","Victima","Edad"]
```

```python
le = LabelEncoder()
```

```python
for col in specific_columns:
    df[col] = le.fit_transform(df[col])
```

```python
df.to_csv("encoded_file.csv", index=False)
```

```python
encoded_df = pd.read_csv("encoded_file.csv")

encoded_df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
# Load the encoded data
encoded_df = pd.read_csv("encoded_file.csv")

encoded_df["Fecha"] = pd.to_datetime(encoded_df["Fecha"]).astype(int)
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(encoded_df)
```

```python
etiquetas = kmeans.predict(encoded_df)
```

```python
centroides = kmeans.cluster_centers_
```

```python
import matplotlib.pyplot as plt

plt.scatter(encoded_df["Delito"], encoded_df["Provincia"], c=etiquetas)

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c="red", marker="x", s=100)

for i, centroid in enumerate(kmeans.cluster_centers_):
    plt.text(centroid[0], centroid[1], f"Cluster {i+1}", fontsize=10)

plt.title("Diagrama de dispersión")
plt.xlabel("Delito")
plt.ylabel("Provincia")
plt.show()
```
